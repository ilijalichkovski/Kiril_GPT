```python
from lxml import etree
import html
from bs4 import BeautifulSoup
import re
import os
import tqdm

print(os.listdir('C:/Users/ilija/Downloads/mk (3)/TED2020/xml/mk'))
```

```python
sample_file_path = f'C:/Users/ilija/Downloads/mk (3)/TED2020/xml/mk/ted2020-1205.xml'

# parse the XML file and get string
tree = etree.parse(sample_file_path)
root = tree.getroot()
str_from_etree = etree.tostring(root, pretty_print=True).decode('utf-8')

# parse HTML and get raw text
soup = BeautifulSoup(str_from_etree, 'html.parser')
raw = soup.get_text()

# remove new lines and excess spaces
fixed = raw.replace('\n', ' ').replace(' .   ', '. ').replace(' , ', ', ').replace(' ?   ', '? ').replace(' : ', ': ')

# remove the brackets that describe laughter, applause etc
no_brackets = re.sub(r'\([^)]+\)', '', fixed)

print(no_brackets)
```

```python
files_list = os.listdir('C:/Users/ilija/Downloads/mk (3)/TED2020/xml/mk')

with open('transcripts_raw.txt', 'w', encoding='utf-8') as f:

    for file in tqdm.tqdm(files_list):
        
        # parse the XML file and get string
        xml_file_path = f'C:/Users/ilija/Downloads/mk (3)/TED2020/xml/mk/{file}'
        tree = etree.parse(xml_file_path)
        root = tree.getroot()
        str_from_etree = etree.tostring(root, pretty_print=True).decode('utf-8')

        # parse HTML and get raw text
        soup = BeautifulSoup(str_from_etree, 'html.parser')
        raw = soup.get_text()

        # remove new lines and excess spaces
        fixed = raw.replace('\n', ' ').replace(' .   ', '. ').replace(' , ', ', ').replace(' ?   ', '? ').replace(' : ', ': ')

        # remove the brackets that describe laughter, applause etc
        no_brackets = re.sub(r'\([^)]+\)', '', fixed)
        
        f.write(no_brackets)
```
